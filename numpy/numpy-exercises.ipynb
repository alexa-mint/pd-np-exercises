```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng

rng = default_rng(42)
mean, stddev = 5, 1
random_normal = rng.normal(mean, stddev, size=500)

plt.figure(figsize=(8, 5))
plt.hist(random_normal, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label=f'Среднее = {mean}')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Гистограмма нормального распределения (n=10)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
